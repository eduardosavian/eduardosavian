Utils

```python
def clip_image(image: np.ndarray) -> np.ndarray:
    image_clipped = np.clip(image, 0, 255)
    return image_clipped

def normalize(image: np.ndarray) -> np.ndarray:
    """Normalize the image and return a numpy array"""
    normalized_image = image / 255.0

    return normalized_image

def denormalize(image: np.ndarray) -> np.ndarray:
    """Denormalize the image and return a numpy array"""
    denormalized_image = (image * 255.0).astype(np.uint8)

    return denormalized_image
```

Convert to grayscale

```python
def convert_to_grayscale_weighted(image: np.ndarray) -> np.ndarray:
    # Dot = multiply the value of the matrix by the RGB pounds
    # R = 0.299
    # B = 0.587
    # G = 0.114
    image_grayscale = np.dot(image[..., :3], [0.299, 0.587, 0.114]).astype(np.uint8)

    return image_grayscale

image_grayscale = convert_to_grayscale_weighted(image_original)
image_grayscale = clip_image(image_grayscale)
cv2_imshow(image_grayscale)
```
